```python
from pyngrok import ngrok
import threading
import os
import time

# Run Streamlit in a separate thread
def run_streamlit():
    os.system(f"streamlit run /content/drive/MyDrive/projects/BoneFracture/FrontEndST/app.py --server.port 8501")  # Updated line

# Start the Streamlit thread
thread = threading.Thread(target=run_streamlit)
thread.start()

# Give Streamlit some time to start
time.sleep(5)

# Open an ngrok tunnel to the Streamlit app
public_url = ngrok.connect(8501).public_url
print("Streamlit app running at:", public_url)
```
